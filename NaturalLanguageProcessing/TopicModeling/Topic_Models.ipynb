## LDA

We will use the `gensim` library for topic modeling with LDA. We import all things necessary to read in the data and transform it into a TF-IDF weighted bag-of-words format

```python
from gensim.models import LdaMulticore, TfidfModel, CoherenceModel
from gensim.corpora import Dictionary # create the dictionary/vocabulary from the data
from gensim.models.phrases import Phrases # join frequent phrases (collocations)

import time # to know how long training took
import multiprocessing # to speed things up by parallelizing
```

We read in the data as a list of lists of strings

```python
instances = df['description_cleaned'].apply(str.split).tolist()
print(instances[0])
```

Now, we can merge all phrases that occur frequently together (e.g., `'new'`, `'york'` -> `'new york'`)

```python
print("merging phrases", flush=True)
phrases = Phrases(instances, min_count=5, threshold=1)
instances_colloc = phrases[instances]
print(instances_colloc[0])

print("creating dictionary", flush=True)
# read in instances and create Dictionary object w information about frequencies etc. 
dictionary = Dictionary(instances_colloc)

# get rid of words that are too rare or too frequent
dictionary.filter_extremes(no_below=50, no_above=0.3)
print(dictionary, flush=True)
```

To work with the data, we need to replace each word with a numerical ID

```python
#replace words by their numerical IDs and their frequency
print("translating corpus to IDs", flush=True)
ldacorpus = [dictionary.doc2bow(text) for text in instances_colloc]
print(ldacorpus[0]) 
```

We now weigh the raw counts for each word in the BOW with its TFIDF value

```python
# learn TFIDF values from corpus
print("tf-idf transformation", flush=True)
tfidfmodel = TfidfModel(ldacorpus)

# transform raw frequencies into TFIDF
model_corpus = tfidfmodel[ldacorpus]
print(model_corpus[0])
```

Our corpus is now ready to run LDA on it.

## Choosing the number of topics

One of the most important parameters is the number of topics. We can select the best value by choosing a subset of the data and measuring various coherence scores for different numbers of topics.

```python
coherence_values = []

dev_size = 15000

for num_topics in range(5, 16):
    model = LdaMulticore(corpus=model_corpus[:dev_size], 
                         id2word=dictionary, 
                         num_topics=num_topics, 
                         random_state=42)

    coherencemodel_umass = CoherenceModel(model=model, 
                                          texts=instances_colloc[:dev_size], 
                                          dictionary=dictionary, 
                                          coherence='u_mass')

    coherencemodel_cv = CoherenceModel(model=model, 
                                       texts=instances_colloc[:dev_size], 
                                       dictionary=dictionary, 
                                       coherence='c_v')

    umass_score = coherencemodel_umass.get_coherence()
    cv_score = coherencemodel_cv.get_coherence()
    
    print(num_topics, umass_score, cv_score)
    coherence_values.append((num_topics, umass_score, cv_score))
```

We can plot the scores to get a better sense of which number of topics to pick

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn
seaborn.set_context('poster') # use large font

scores = pd.DataFrame(coherence_values, columns=['num_topics', 'UMass', 'CV'])

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
scores.plot.line(x='num_topics', y='UMass', ax=ax[0], xticks=range(5,16));
scores.plot.line(x='num_topics', y='CV', ax=ax[1], xticks=range(5,16));
```

With this visual aid, we can pick our main parameter and train a model.

```python
num_topics = 10

# find chunksize to make about 200 updates
num_passes = 10
chunk_size = len(model_corpus) * num_passes/200
print(chunk_size)

start = time.time()
print("fitting model", flush=True)
model = LdaMulticore(num_topics=num_topics, # number of topics
                     corpus=model_corpus, # what to train on 
                     id2word=dictionary, # mapping from IDs to words
                     workers=min(10, multiprocessing.cpu_count()-1), # choose 10 cores, or whatever computer has
                     passes=num_passes, # make this many passes over data
                     chunksize=chunk_size, # update after this many instances
                     alpha=0.5,
                     random_state=42
                    )
    
print("done in {}".format(time.time()-start), flush=True)
```

With the fitted model, we can transform each instance into a distribution over topics:

```python
# transform the data into topic distros
topic_corpus = model[model_corpus]

topic_corpus[0]
```

We might also want to know what the topics are:

```python
model.print_topics()
```

Not very readable: the word prevalence and the word are all in one long string. Let's use regular expressions to chop it up a bit to make it more readable.

```python
import re

# get the topic descriptions
topic_sep = re.compile("0\.[0-9]{3}\*") # getting rid of useless formatting
# extract a list of tuples with topic number and descriptors from the model
model_topics = [(topic_no, re.sub(topic_sep, '', model_topic).split(' + ')) for topic_no, model_topic in
                model.print_topics(num_topics=num_topics, num_words=5)]

descriptors = []
for i, m in model_topics:
    print(i+1, ", ".join(m[:5]))
    descriptors.append(", ".join(m[:2]).replace('"', ''))
```

Ok, that looks a bit more like it! Seems like we have identified a number of wine profiles

## Aggregating topics by a dependent variable

If our topics are indeed wine profiles, then it makes sense that each country has their own mix of profiles: Germany produces mainly white wines, whereas the others produce both, but with different emphasis. Can we split things by country?

```python
target_category = 'country'
# get a list of all the topic scores for each document
scores = [[t[1] for t in topic_corpus[entry]] for entry in range(len(topic_corpus))]

# turn that into a data frame with N rows and K columns, each with the score of the corresponding topic
topic_distros = pd.DataFrame(data=scores, columns=descriptors)

# add the review category of each document (so we can aggregate)
topic_distros['category'] = df[target_category]
```

```python
topic_distros.head()
```

Let's turn that into a graph

```python
seaborn.set_context('poster') # use large font
fig, ax = plt.subplots(figsize=(20, 10)) # set graph size

# aggregate topics by categories
aggregate_by_category = topic_distros.groupby(topic_distros.category).mean()

# plot the graph
aggregate_by_category[descriptors].plot.bar(ax=ax);

# move the legend out
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5));
```

# Author Topic Model

Aggregating the topics by a covariate like country is nice, but instead of doing it after the fact, we can also build it into the modeling process itself. The Author Topic Model learns a different topic distribution for each covariate. As a result, the distributions are much sparser than when aggregating.

```python
from gensim.models import AuthorTopicModel
from gensim.test.utils import datapath, temporary_file
```

Importantly, we need to assign all documents to a covariate (again, country) via a dictionary. We also need to throw out empty instances, and for some reason, we have to use raw counts instead of TFIDF weighted ones.

```python
from collections import defaultdict

# mapping from country to document ID
author2doc = defaultdict(list)

i = 0

author_model_corpus = []
author_instances = []
for doc, tfidf_doc, country in zip(instances_colloc, ldacorpus, df.country):
    if len(tfidf_doc) > 0:
        author_model_corpus.append(tfidf_doc)
        author2doc[country].append(i)
        author_instances.append(doc)
        i += 1
    
len(author2doc), len(author_model_corpus), len(author_instances)
```

The result is a mapping of country to a list of document IDs (i.e., a list of integers)

```python
author2doc['Italy'][:10]
```

Just as before, we need to choose a number of topics. Because we use a different model structure, we should rerun the selection.

```python
coherence_values = []
author_model_list = []

dev_size = 10000

dev_author2doc = {key: [idx for idx in value if idx < dev_size] for key, value in author2doc.items()}

for num_topics in range(5, 11):
    author_model = AuthorTopicModel(corpus=author_model_corpus[:dev_size], 
                                    author2doc=dev_author2doc, 
                                    id2word=dictionary, 
                                    num_topics=num_topics,
                                    random_state=42
                                   )
    
    coherencemodel_umass = CoherenceModel(model=author_model, 
                                          texts=author_instances[:dev_size], 
                                          dictionary=dictionary, 
                                          coherence='u_mass')

    coherencemodel_cv = CoherenceModel(model=author_model, 
                                       texts=author_instances[:dev_size], 
                                       dictionary=dictionary, 
                                       coherence='c_v')

    umass_score = coherencemodel_umass.get_coherence()
    cv_score = coherencemodel_cv.get_coherence()
    
    print(num_topics, umass_score, cv_score)
    coherence_values.append((num_topics, umass_score, cv_score))
```

As before, let's plot it to get a sense what to choose.

```python
%matplotlib inline
scores = pd.DataFrame(coherence_values, columns=['num_topics', 'UMass', 'CV'])
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
scores.plot.line(x='num_topics', y='UMass', ax=ax[0], xticks=range(5,16));
scores.plot.line(x='num_topics', y='CV', ax=ax[1], xticks=range(5,16));
```

From here on, it is the exact same procedure as before

```python
n_topics_author = 5

author_model = AuthorTopicModel(corpus=author_model_corpus, 
                                author2doc=author2doc, 
                                id2word=dictionary, 
                                num_topics=n_topics_author,
                                passes=num_passes,
                                chunksize=chunk_size,
                                random_state=42
)
```

The return format is slightly different, so to extract the topic descriptors, we use a different code

```python
# extract a list of tuples with topic number and descriptors from the model
author_model_topics = [(topic_no, re.sub(topic_sep, '', model_topic).split(' + ')) for topic_no, model_topic in
                author_model.print_topics(num_topics=n_topics_author, num_words=5)]

author_descriptors = []
for i, m in sorted(author_model_topics):
    print(i+1, ", ".join(m[:5]))
    author_descriptors.append(", ".join(m[:2]).replace('"', ''))
```

We can now create the distributions over topics for each country. Since some of them will be 0 (not all countries have all topics), we need to explicictly initialize the distribution 

```python
# initialize mapping from covariate/author/country to topic distro, set all to 0.0
author_vecs = {author: {author_descriptors[t]: 0.0
                         for t in range(author_model.num_topics)}
              for author in author_model.id2author.values()
              }
# update mappings from model
for author in author_model.id2author.values():
    for (t, v) in author_model.get_author_topics(author):
        author_vecs[author][author_descriptors[t]] = v

for country in 'Germany US Italy France Spain'.split():
    print(country, author_vecs[country])        
```

All that is left to do is plot it

```python
# make a DataFrame
author_df = pd.DataFrame.from_dict(author_vecs)

# plot it
fig, ax = plt.subplots(figsize=(20,10))
author_df.T.plot.bar(ax=ax)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5));
```
